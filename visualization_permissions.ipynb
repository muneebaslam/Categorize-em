```python
# plotting a bipartite graph two sets of nodes one contains the apps and other contains the permissions required 
import networkx as nx
from matplotlib import pyplot as plt

node_list = nx.read_edgelist('C:/Users/Muneeb/Desktop/PakisaniAppsPermissionData.txt')
nx.draw(node_list,node_size = 10,with_labels = False,font_size= 5,width = 1,node_color = 'blue',pos = nx.spring_layout(node_list))
plt.show()
```
