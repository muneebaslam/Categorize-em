```python
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.read_edgelist('C:/Users/Muneeb/Desktop/Al-Data/social.txt')

color_map = []
print(nx.info(graph))
for node in graph:
    if node == 'Social':
        color_map.append('red')
    else:
        color_map.append('blue')
nx.draw(graph,with_labels = True,font_size = 10,node_size = 60,node_color = color_map,pos = nx.spring_layout(graph),k=45)
plt.show()
```
